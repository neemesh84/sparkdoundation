```python
X = df.drop("Scores",axis=1).values #Feature/independent variable
y = df['Scores'].values #target/dependent variables
```

# Regression Analysis 

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
prediction_space = np.linspace(min(X),max(X)).reshape(-1, 1)
```

A linear regression line has an equation of the form Y = mX+c, where X is the explanatory variable and Y is the dependent variable. The slope of the line is m, and c is the intercept or slopes (the value of y when x = 0).

```python
plt.scatter(X, y, color='red')
plt.plot(prediction_space, reg.predict(prediction_space),
color='blue', linewidth=4)
plt.show()
```

# Train the model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python

fit=reg.fit(X_train,y_train)
```

```python
print("Coefficient:     ",fit.coef_)     # Coefficient 
print("Intercept: ",fit.intercept_)  # intercept
```
